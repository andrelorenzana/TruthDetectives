```python
# Dependencies
from bs4 import BeautifulSoup
import requests
import pymongo
import numpy as np
```

```python
# Initialize PyMongo to work with MongoDBs
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)
```

```python
# Define database and collection
db = client.snopesfinal_db 
collection = db.false_articles_final
```

```python
for n in range(1,301):
    # URL of page to be scraped
    #false_url = 'https://www.snopes.com/fact-check/rating/false/'
    false_url_long = 'https://www.snopes.com/fact-check/rating/false/page/%d/' % (n)
    ##soup = BeautifulSoup(false_url_long, 'html.parser')
    #print('-----------------')
    #print(false_url_long)
    # Retrieve page with the requests module
    f_response = requests.get(false_url_long)
    # Create BeautifulSoup object; parse with 'lxml'
    false_soup = BeautifulSoup(f_response.text, 'lxml')
    #print(false_soup)
    # Retrieve the parent divs for all articles
    f_results = false_soup.find_all('div', class_='media-body')

    # Loop through results to retrieve items
    for result in f_results:
        try:
            title = result.find('h2', class_='card-title').text
            # The time and date of article publication
            #date = result.find('span', class_='date').text
            #print(title)

            # Dictionary to be inserted into MongoDB
            post = {
                'Title': title,
                'Text': np.nan,
                'Author': np.nan,
                'Source': "Snopes Aggregator",
                'Label': "Unreliable"
                    }

            # Insert dictionary into MongoDB as a document
            collection.insert_one(post)
        
        except Exception as e:
            print(e)
```
