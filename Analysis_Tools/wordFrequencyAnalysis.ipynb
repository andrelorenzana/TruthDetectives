```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Loading Data

```python
# Only works for csv files
def load_data(file_path):
    df = pd.read_csv(file_path)
    
    if 'Unnamed: 0' in df.columns:
        df = df.drop('Unnamed: 0', axis=1)
    
    return df
```

```python
reliable_title_df = load_data('../Analysis_Datasets/reliable_title_word.csv')
unreliable_title_df = load_data('../Analysis_Datasets/unreliable_title_word.csv')
satirical_title_df = load_data('../Analysis_Datasets/satirical_title_word.csv')
reliable_text_df = load_data('../Analysis_Datasets/reliable_text_word.csv')
unreliable_text_df = load_data('../Analysis_Datasets/unreliable_text_word.csv')
```

# Title Analysis
## Reliable Sources

```python
sns.set_style("darkgrid")
sns.set(font_scale = 3.2)
plt.figure(figsize=(28,15))
plot = sns.barplot(x='Word', y='Frequency', data = reliable_title_df.head(50), palette="Blues_d")
plt.setp(plot.get_xticklabels(), rotation=90)
plt.title('Most Frequent Title Words from Reliable Sources')
plt.tight_layout()
plt.savefig('../Figures/reliable_title_words')
plt.show()
```

## Unreliable Sources

```python
sns.set_style("darkgrid")
sns.set(font_scale = 3.2)
plt.figure(figsize=(28,15))
plot = sns.barplot(x='Word', y='Frequency', data = unreliable_title_df.head(50), palette="Blues_d")
plt.setp(plot.get_xticklabels(), rotation=90)
plt.title('Most Frequent Title Words from Unreliable Sources')
plt.tight_layout()
plt.savefig('../Figures/unreliable_title_words')
plt.show()
```

## Satirical Sources

```python
sns.set_style("darkgrid")
sns.set(font_scale = 3.2)
plt.figure(figsize=(28,15))
plot = sns.barplot(x='Word', y='Frequency', data = satirical_title_df.head(50), palette="Blues_d")
plt.setp(plot.get_xticklabels(), rotation=90)
plt.title('Most Frequent Title Words from Satirical Sources')
plt.tight_layout()
plt.savefig('../Figures/satirical_title_words')
plt.show()
```
